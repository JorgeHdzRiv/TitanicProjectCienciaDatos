# Introduccion

El dataset del Titanic es uno de los conjuntos de datos más conocidos y ampliamente utilizados en ciencia de datos y aprendizaje automático. Este dataset contiene información sobre los pasajeros del Titanic, el famoso transatlántico que se hundió en su viaje inaugural en 1912

![](../img/header.png)

## Descripcion de los datos

| Columna      | Descripción                                         | Notas                                                                                       |
|--------------|-----------------------------------------------------|---------------------------------------------------------------------------------------------|
| PassengerId  | Identificador único para cada pasajero              |                                                                                             |
| Survived     | Indica si el pasajero sobrevivió (1) o no (0)       |                                                                                             |
| Pclass       | Clase del boleto del pasajero (1 = Primera, 2 = Segunda, 3 = Tercera) | Una proxy para el estatus socioeconómico (SES). 1st = Alta, 2nd = Media, 3rd = Baja        |
| Name         | Nombre del pasajero                                 |                                                                                             |
| Sex          | Género del pasajero                                 |                                                                                             |
| Age          | Edad del pasajero                                   | La edad es fraccional si es menor a 1. Si la edad es estimada, está en la forma de xx.     |
| SibSp        | Número de hermanos y cónyuges a bordo del Titanic   | El dataset define las relaciones familiares de esta manera: Hermano = hermano, hermana, hermanastro, hermanastra. Cónyuge = esposo, esposa (se ignoraron amantes y prometidos) |
| Parch        | Número de padres e hijos a bordo del Titanic        | El dataset define las relaciones familiares de esta manera: Padre = madre, padre. Hijo = hija, hijo, hijastra, hijastro. Algunos niños viajaban solo con una niñera, por lo tanto parch=0 para ellos. |
| Ticket       | Número del boleto                                   |                                                                                             |
| Fare         | Tarifa pagada por el boleto                         |                                                                                             |
| Cabin        | Número de cabina                                    |                                                                                             |
| Embarked     | Puerto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton) |                                                                                             |
